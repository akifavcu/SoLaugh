```python
# Import dataframe
df = pd.read_csv('/run/media/clara_elk/Clara_Seagate/SoLaugh_MSc/results/behavior/active_task/subj-all_task-LaughterActive.csv')
```

```python
count_correct = []
for i, elt in enumerate(df['response']) :
    if 'correct' in elt:
        count_correct.append(1)
    else :
        count_correct.append(0)

df['count_correct'] = count_correct
```

```python
# Remove None
df_clean = df[df['response'].str.contains('none') == False]
```

## Nombre de stimuli par locuteur

```python
df_nb_speaker = df_clean[['fileName','speaker']].groupby(['fileName', 'speaker'], as_index=False).count()
```

```python
ces,sop,sam,cat,tob,dan = 0, 0, 0, 0, 0, 0

for elt in df_nb_speaker['speaker'] :
    if elt == 'ces':
        ces += 1
    elif elt == 'tob' :
        tob += 1 
    elif elt == 'sam' :
        sam += 1
    elif elt == 'cat' :
        cat += 1 
    elif elt == 'dan' :
        dan += 1
    elif elt == 'sop' :
        sop += 1

dict_speaker = {'speaker' : ['ces','sop','sam','cat','tob','dan'],
                'nb_stim' : [ces,sop,sam,cat,tob,dan]}

df_speaker_stim = pd.DataFrame.from_dict(dict_speaker, orient='columns')
df_speaker_stim
```

```python
ax = sns.barplot(data=df_speaker_stim, x='speaker', y='nb_stim', order=['sop', 'cat', 'tob', 'sam', 'ces', 'dan'], palette='pastel')
ax.set(xlabel='Locuteurs', ylabel='Nombre de stimuli', title='Nombre de stimuli par locuteur')

plt.savefig(BEHAV_PATH + 'subj-all_task-LaughterActive_plot-stimuli-per-speaker.pdf')
plt.savefig(BEHAV_PATH + 'subj-all_task-LaughterActive_pot-stimuli-per-speaker.png')
```
